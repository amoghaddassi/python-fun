Data sources:

MLB Game Log: https://data.world/dataquest/mlb-game-logs

NYY Attendance: https://www.baseball-reference.com/teams/NYY/attend.shtml

NYM Attendance: https://www.baseball-reference.com/teams/NYM/attend.shtml

```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def to_int(string_list):
    ints = []
    for string in string_list:
        str_list = list(string)
        ints.append( int(''.join([x for x in str_list if x !=','])) )
    return ints
```

```python
game_log = pd.read_csv('game_logs.csv');

game_log = game_log[['date', 'h_name', 'v_name', 'h_score', 'v_score']]
game_log['h_win'] = game_log['h_score'] > game_log['v_score']
game_log['v_win'] = game_log['h_score'] < game_log['v_score']
years = []
for date in game_log['date']:
    years.append(int(str(date)[:4]))
game_log['year'] = years

nyy_home = game_log[(game_log['h_name'] == 'NYA') & (game_log['year'] >= 1962)].reset_index().drop('index', axis=1)
nyy_away = game_log[(game_log['v_name'] == 'NYA') & (game_log['year'] >= 1962)].reset_index().drop('index', axis=1)

nym_home = game_log[game_log['h_name'] == 'NYN'].reset_index().drop('index', axis=1)
nym_away = game_log[game_log['v_name'] == 'NYN'].reset_index().drop('index', axis=1)

nyy_home_wins = nyy_home.groupby('year')['h_win'].agg(np.count_nonzero)
nyy_away_wins = nyy_away.groupby('year')['v_win'].agg(np.count_nonzero)
nyy_home_pct = nyy_home_wins / (nyy_home_wins + nyy_away_wins)
nyy_runs = nyy_home.groupby('year')['h_score'].agg(np.average)

nym_home_wins = nym_home.groupby('year')['h_win'].agg(np.count_nonzero)
nym_away_wins = nym_away.groupby('year')['v_win'].agg(np.count_nonzero)
nym_home_pct = nym_home_wins / (nym_home_wins + nym_away_wins)
nym_runs = nym_home.groupby('year')['h_score'].agg(np.average)
```

```python
nyy_attendance = pd.read_csv('nyy_attendance.csv')
nyy_attendance = nyy_attendance[nyy_attendance['Year'] >= 1962].set_index('Year')['Attend/G']

nym_attendance = pd.read_csv('nym_attendance.csv')
nym_attendance = nym_attendance[nym_attendance['Year'] >= 1962].set_index('Year')['Attend/G']
```

```python
plt.figure()
plt.scatter(to_int(nyy_attendance[nyy_home_pct.index]), nyy_runs, c=nyy_home_pct.index, cmap='viridis')
plt.scatter(to_int(nym_attendance[nym_home_pct.index]), nym_runs, c=nym_home_pct.index, cmap='viridis')
plt.colorbar()
plt.xlabel('Average Attendance per Game')
plt.ylabel('Average Runs Scored by NY Team')
plt.title('NY Baseball Homegames from 1962-2016');
```
